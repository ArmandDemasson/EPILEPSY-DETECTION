```python
import numpy as np
import pandas as pd
import os.path as op
from sklearn.feature_selection import SelectKBest, chi2
import pickle
from scipy.fft import fft, fftfreq
from tqdm import tqdm
```

```python
def load_obj(name, path):
    with open(op.join(path, name), 'rb') as f:
        return pickle.load(f)
```

```python
results_df = pd.DataFrame(load_obj("modelling_agregated_feature.pkl","/sps/crnl/ademasson/data/features_extraction"))
results_df
```

```python
test_columns = [col for col in results_df.columns.tolist() if col.startswith('agr_test_')]
test_columns
```

```python
pred_columns = [col for col in results_df.columns.tolist() if col.startswith('agr_pred_')]
pred_columns
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
ncv = 1
for test, pred in zip(test_columns, pred_columns):
    print(f"CV {ncv}")
    print(confusion_matrix(results_df[test], results_df[pred]))
    print(classification_report(results_df[test], results_df[pred]))
    ncv +=1
```
