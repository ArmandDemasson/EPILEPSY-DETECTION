```python
import numpy as np
import pandas as pd
import os.path as op
from sklearn.feature_selection import SelectKBest, chi2
import pickle
from scipy.fft import fft, fftfreq
from tqdm import tqdm
```

```python
def load_obj(name, path):
    with open(op.join(path, name), 'rb') as f:
        return pickle.load(f)
```

```python
results_df = pd.DataFrame(load_obj("modelling_8_feature.pkl","/sps/crnl/ademasson/data/features_extraction"))
results_df
```

```python
y= {}
for i in range(1,6):
    y[f'f_{str(i)}'] = {f'cv_{str(j+1)}': {"test": results_df[f'f_{str(i)}_test_cv_{str(j)}'], "pred": results_df[f'f_{str(i)}_pred_cv_{str(j)}']} for j in range(0,5)}
```

```python
y[f'f_{str(8)}'] = {f'cv_{str(j+1)}': {"test": results_df[f'f_{str(8)}_test_cv_{str(j)}'], "pred": results_df[f'f_{str(8)}_pred_cv_{str(j)}']} for j in range(0,5)}
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
for f, cv in y.items():
    print(f"-----{f[-1]} features-----")
    for ncv, res in cv.items():
        print(f"CV {ncv[-1]}")
        print(confusion_matrix(res['test'], res['pred']))
        print(classification_report(res['test'], res['pred']))
```
